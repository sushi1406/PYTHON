```python
ports=pd.get_dummies(titanic_data.Embarked, prefix="Embarked",dtype=int)
```

```python
ports.head()
```

```python
titanic_data=titanic_data.join(ports)
```

```python
titanic_data.drop(['Embarked'],axis=1, inplace=True)
```

```python
titanic_data.head(5)
```

```python
titanic_data.Sex=titanic_data.Sex.map({'male':0, 'female':1})
```

```python
titanic_data['Sex']
```

```python
sns.countplot(x=titanic_data['Sex'],hue=titanic_data['Survived'])
plt.xlabel('Sex')
plt.ylabel('Count')
plt.title('Survival Count by Gender')
plt.legend(title='Survived', loc='upper right')
plt.show()
```

```python
#Splitting the data
Y=titanic_data.Survived.copy()
X=titanic_data.drop(['Survived'],axis=1)
```

```python
sns.heatmap(titanic_data.corr(numeric_only = True), cmap='YlGnBu')
plt.show
```

```python
X.drop(['Cabin','Name','PassengerId','Ticket'],axis=1,inplace=True)
```

```python
#checking the null values
X.isnull().sum()
```

```python
#filling the null values in Age with the mean Age
X["Age"] = X["Age"].fillna(X["Age"].mean())
```

```python
X.isnull().sum().values.any()
```

```python
#importing the libraries for model building
from sklearn.model_selection import train_test_split
```

```python
#splitting the data to train our model and test its accuracy
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=7)
```

```python
#import regression libraries for model building
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
#Fitting our data to the model
model.fit(X_train,Y_train)
```

```python
model.score(X_train,Y_train)
```

```python
#predict on the model
Y_pred_test=model.predict(X_test)
Y_pred_test
```

```python
Y_test
```

```python
#import metrics to check accuracy
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(Y_test,Y_pred_test)
print('Model accuracy score: {:.4f}'.format(accuracy))
```
